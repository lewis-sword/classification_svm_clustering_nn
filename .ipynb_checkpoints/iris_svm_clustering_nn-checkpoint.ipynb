# K-means clustering vs. Neural network vs. Support vector machine learning example #
Here we apply K-means clustering to the iris dataset from pydataset, using sklearn. We can extract the inertia quantity and apply the elbow method
to quantify the number of clusters we should use. We then construct a simple MLP classifier to test the classifications. Finally we use a support vector machine. The results are collected and compared to the true results at the end.

# Clustering #

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pydataset import data
import tensorflow as tf
import sklearn
from sklearn.cluster import KMeans
```

```python
datasets = data()
print(datasets)
```

We'll use the 'iris' dataset. show_doc=True for more information

```python
#data('iris', show_doc=True)
data('iris')
df = data('iris')
print(df.head())
```

```python
df.columns
```

```python
df["Species"].unique()
```

**Make a scatter plot of the petal lengths against petal widths and colour the points by the true class label (species)**

```python
def scatter_def(i, colour):
    return plt.scatter(df[df["Species"]==i]["Petal.Width"],df[df["Species"]==i]["Petal.Length"], color=colour)

for i, species in enumerate(df["Species"].unique()):
    print([i, species])

plt.figure()
scatter_def("setosa","blue")
scatter_def("versicolor","red")
scatter_def("virginica","green")
plt.show()
```

```python
# Using the data from the 4 remaining features, we want to cluster the data and compar to the target feature "Species" of flower.
# Set a new dataframe to do so
df_y = df[['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']]
df_y.head()
```

```python
#####Elbows#####
max_cluster=10
no_of_clust_list = [i for i in range(1,max_cluster)]
inertia_list = []
for i in range(1,max_cluster):
    km_norm = KMeans(n_clusters=i)
    y_km_norm = km_norm.fit_predict(df_y)
    inertia_list.append(km_norm.inertia_)
plt.plot(no_of_clust_list,inertia_list)
```

**From inertia plot, take K=3, and use the KMeans clustering algorithm to find clusters.**

```python
km = KMeans(n_clusters=3)
y_km = km.fit_predict(df_y)
# The predicted class integers from the clustering algorithm are printed
print(y_km)
```

**Two subplots of petal widths against petal lengths. Each features colours according to classification.**

```python
#Note we can use the original dataframe df here, so long as the colour scheme is generated by y_km: the output of fitting on 
# the df_y which does not contain species feature data.
plt.figure()

#Cluster
plt.subplot(1,2,1)
plt.scatter(df["Petal.Width"],df["Petal.Length"], c= y_km)
plt.title("Cluster")

#True
plt.subplot(1,2,2)
scatter_def("setosa","orange")
scatter_def("versicolor","green")
scatter_def("virginica","blue")
plt.title("True")
```

<font color="blue">

**Overlapping clusters?**
Note that when the number of input features for the kmeans algorithm is equal to 2, call them (x,y), we can only achieve clusters which have clear seperations in the x-y plane. That is data points belonging to one cluster will be completely separate to another cluster. However when we have 3 (or more) input features, we essentially have another dimension to work with which means that instead of setting lines in a plane to separate the clusters, we have 2d surfaces in 3-dimenisons that seperate the clusters (or the higher dim generalisation).

# Creating a neural network to predict species # 

```python
#df_for_km puts the 4 independent features into an array
df_x_data=df.iloc[:, [0,1,2, 3]].values
#mapping the species to separate integers, then creating a 1 column dataframe for them
mapping = {"setosa": 0, "versicolor": 1, "virginica":2}
df_y_data = df.replace({'Species': mapping}).iloc[:, [4]].values
x_data=np.array(df_x_data)
y_data=np.array(df_y_data)
#print(y_data)

#Splitting into train and test data
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(x_data, y_data, test_size=0.3)
print(X_train.shape, y_train.shape,np.zeros((10, 4)).shape)
```

```python
ValidationSplit = 0.35
BatchSize       = 10
Nepochs         = 200
DropoutValue    = 0.25


#Number of nodes/layers probably overkill for the size of the dataset
model = tf.keras.models.Sequential()
# takes the shape of the dataframe, in case we want to change the number of features.
model.add(tf.keras.layers.Dense(64, input_shape=(df_x_data.shape[1],), activation="relu"))
model.add(tf.keras.layers.Dropout(DropoutValue))
model.add(tf.keras.layers.Dense(128, activation="relu"))
model.add(tf.keras.layers.Dropout(DropoutValue))
model.add(tf.keras.layers.Dense(256, activation="relu"))
model.add(tf.keras.layers.Dropout(DropoutValue))
#final layer has 3 nodes for each of the potential species
model.add(tf.keras.layers.Dense(3, activation='softmax'))

loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
learning_rate = 0.00008 # alpha.

model.compile(optimizer=tf.optimizers.Adam(learning_rate=learning_rate), loss=loss_fn,metrics=['accuracy'])
history  = model.fit(X_train, y_train, validation_split=ValidationSplit, batch_size=BatchSize, epochs=Nepochs)
model.summary()
```

**Seeing how the data performs at classifying the unseen, test data sets**

```python
#print(len(pred),len(y_test))

#One can make model predictions using reshaping or simply calling on numpy array to do so.
model.predict(np.array(X_test))
#To get the class labels we find the maximum value in each array row (these rows with 3 values correspond to the three types of flower categories 0,1,2)
#and obtain the max entry in each row with argmax. This tells us the species prediction for that data row.
#For example
pred= np.argmax(model.predict(np.array(X_test)), axis=1)
# print(pred)
# print(np.array([y_test[a][0] for a in range(len(y_test))]))

# Testing which predicted labels match true labels
count=0
res_list=[]
for i in range(len(pred)):
    test = pred[i]== y_test[i];
    res_list.append(test[0])
    if pred[i]== y_test[i]:
        count +=1
ratio_val = float(count)/len(pred)
#print(res_list)
print("Successful prediction percentage (NN): ", ratio_val)
```

**Make plots of the loss and accuracy for the training and validation sets**

```python
plt.figure()
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Training and Validation Loss")
plt.clf

plt.figure()
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Training and Validation Accuracy")
plt.clf
#print("Validation accuracy = ", history.history["val_accuracy"])
```

The final epoch still shows no overfitting and generally we have a large increase in accuracy. **Warning** the data set is very small and so the results may not generalise well.

**Plotting the results of all the data evaluated on the neural network model.**

```python
#Realistically the model has already seen the majority of the x_data in training and validation. However, we can still use the model/weights
#to make a prediction on it.
full_data_pred=model.predict(np.array(x_data))
pred2 = np.argmax(full_data_pred, axis=1)
print(pred2)

plt.figure(figsize=(12,7))
#NN
plt.subplot(1,4,1)
plt.scatter(df["Petal.Width"],df["Petal.Length"], c= pred2)
plt.title("NN")
```

# Support vector machine #

```python
# Short additional code for classifying with a support vector machine. Note, the feature space used is 4 dimensional, which also means that even using a 
# linear kernel, we can still expect separation from the two, non-linearly separable species classes. "linear" in a 4 dimensional feature space still 
# allows for a manifold which could identify the three classes. "poly" in a 2 dimensional feature space should still permit individual classification, 
# given the projection to a higher space would allow this. "linear" in a 2d feature space (like the plots, petal width vs. petal length) would imply 
# using one line to separate the groups.

# What's most important: we can always insert more than one, 1d dimensional boundary, which is why, even when predicting using two features,
# the SVM process plots two, linear boundary lines for a linear kernel. The caveat is, that while one of those lines perfectly separates one group 
# from the other two, the other line(s), no matter where it is (they are) put, cannot separate the remaining two groups. This is why the polynomial
# option is important: its boundaries can be nonlinear in the (feature1, feature2) space, hence can separate the remaining two groups better.

from sklearn.svm import SVC # "Support vector classifier"
model = SVC(kernel='linear', C=1E10)
svm_fit=model.fit(X_train, y_train.reshape(len(y_train),))

svm_pred=model.predict(X_test)

# Testing which predicted labels match true labels
count_svm=0
res_list_svm=[]
for i in range(len(svm_pred)):
    test = svm_pred[i]== y_test[i];
    res_list_svm.append(test[0])
    if svm_pred[i]== y_test[i]:
        count_svm +=1
ratio_val = float(count_svm)/len(svm_pred)
#print(res_list)
print("Successful prediction percentage (SVM): ", ratio_val)

svm_pred_full = model.predict(x_data)
#SVM
plt.figure(figsize=(12,6))
plt.subplot(1,4,3)
plt.scatter(df["Petal.Width"],df["Petal.Length"], c= svm_pred_full)
plt.title("SVM")
plt.show()
```

**Generally, all three methods perform well, with very few outliers compared to the true plot**

## SVM testing with lower dimensional feature space ##

```python
# Testing the results using only 2 features for the x data: linear kernel can never get a perfect classification as two classes aren't linearly separable 
# and the feature space is not high enough in dimension. Polynomial on the other hand can do the job much better

from mlxtend.plotting import plot_decision_regions

#df_for_km puts the 4 independent features into an array
df_x_data=df.iloc[:, [2,3]].values
#mapping the species to separate integers, then creating a 1 column dataframe for them
mapping = {"setosa": 0, "versicolor": 1, "virginica":2}
df_y_data = df.replace({'Species': mapping}).iloc[:, [4]].values
x_data=np.array(df_x_data)
y_data=np.array(df_y_data)
#print(y_data)

#Splitting into train and test data
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(x_data, y_data, test_size=0.3)

from sklearn.svm import SVC # "Support vector classifier"
model = SVC(kernel='poly', C=1E10) # Change this from linear to poly: with 2features used in training, only poly can get a good fit.
svm_fit=model.fit(X_train, y_train.reshape(len(y_train),))
svm_pred = model.predict(x_data)
print(svm_pred)
#SVM
#plt.scatter(df["Petal.Width"],df["Petal.Length"], c= svm_pred)

plot_decision_regions(X_train, y_train.reshape(len(y_train),), clf=svm_fit, legend=4)
plt.title("SVM")
plt.show()
```
